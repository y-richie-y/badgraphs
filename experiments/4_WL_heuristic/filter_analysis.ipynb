```python
import numpy as np
import matplotlib.pyplot as plt

f1 = open("data/sat_gi.txt")
l1 = f1.readlines()
f2 = open("data/sat_gi2.txt")
l2 = f2.readlines()

def parse_line(s):
    s = s.replace("\n", "").split(" ")
    return [float(s[x]) for x in range(1, 5)]

data1 = np.array(map(parse_line, l1))
data2 = np.array(map(parse_line, l2))

fig = plt.figure(num=None)
ax = fig.add_subplot(111)

ax.set_xlabel("Difference in SAT solve time (s)")
ax.set_ylabel("Automorphism Solve Time (s)")

ax.scatter(data1[:,2], data1[:,3], facecolor="None", edgecolor="b")
plt.show()

fig.savefig("exports/filter_w_filter.pdf", bbox_inches='tight')

fig = plt.figure(num=None)
ax = fig.add_subplot(111)

ax.set_xlabel("Difference in SAT solve time (s)")
ax.set_ylabel("Automorphism Solve Time (s)")

ax.scatter(data2[:,2], data2[:,3], facecolor="None", edgecolor="b")
plt.show()

fig.savefig("exports/filter_wo_filter.pdf", bbox_inches='tight')
```
