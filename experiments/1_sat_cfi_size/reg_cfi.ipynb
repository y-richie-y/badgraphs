```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
matplotlib.rcParams['font.size'] = 13
matplotlib.rcParams['lines.markersize'] = 7
matplotlib.rcParams['lines.linewidth'] = 1
matplotlib.rcParams['figure.facecolor'] = 'w'
matplotlib.rcParams['figure.dpi'] = 80
matplotlib.rcParams['axes.grid'] = 'true'

# extracted using
# awk '/grpsize/ {print "size: "$1} /seconds/ {print "time: "$4}' */nauty_times.txt

f = open("data/nauty.txt")
lines = f.read().split("\n")[:-1]

sizes = list(map(lambda s: int(s.replace("size: ", "")), lines[0::2]))
times = list(map(lambda s: float(s.replace("time: ", "")), lines[1::2]))

fig = plt.figure(num=None, figsize=(9, 6))
ax = fig.add_subplot(111)

ax.scatter(sizes, times, marker="o", facecolors='None', edgecolors="purple")
ax.set_title("reg_cfi's automorphism Solve Time on nauty")
ax.set_ylabel("Solve time (s)")
ax.set_xlabel("Number of Nodes")
plt.show()

fig.savefig("exports/reg_cfi_nauty_times.pdf", bbox_inches='tight')

pairs = sorted([(sizes[i], times[i]) for i in range(len(times))])
```
