```python
import seaborn as sns, numpy as np

f = open("data/distribution.txt")
lines = f.read().split("\n")[:-1]

sizes = list(map(lambda s: int(s.replace("size: ", "")), lines[0::2]))
times = list(map(lambda s: float(s.replace("time: ", "")), lines[1::2]))

sns.set();
ax = sns.distplot(times)
ax.set_title("Distribution of sat_cfi with m=1200 n=600 on nauty")
ax.set_ylabel("Density")
ax.set_xlabel("Time (s)")
ax.set_yticklabels([])
ax.set_xlim(-100, None)

fig = ax.get_figure()
fig.savefig("exports/sat_cfi_dist_2_1_seaborn.pdf", bbox_inches='tight') 
```
