```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
matplotlib.rcParams['font.size'] = 13
matplotlib.rcParams['lines.markersize'] = 7
matplotlib.rcParams['lines.linewidth'] = 1
matplotlib.rcParams['figure.facecolor'] = 'w'
matplotlib.rcParams['figure.dpi'] = 80

f1 = open("data/distribution.txt")
f2 = open("data/with_filter.txt")
f3 = open("data/without_filter.txt")

def make_graph(f, description):
    lines = f.read().split("\n")[:-1]

    sizes = list(map(lambda s: int(s.replace("size: ", "")), lines[0::2]))
    times = list(map(lambda s: float(s.replace("time: ", "")), lines[1::2]))

    median = np.median(times)

    fig = plt.figure(num=None, figsize=(7.5, 5))
    ax = fig.add_subplot(111)

    ax.axvline(x=median, color="green")
    ax.hist(times, bins=60)
    ax.set_title("Distribution of sat_cfi on nauty (%s)" % description)
    ax.set_xlabel("Solve Time (s)")
    ax.set_ylabel("Number of Graphs (per bin)")
    plt.show()

# make_graph(f1, "Collection A")
make_graph(f2, "Collection B1")
make_graph(f3, "Collection B2")
```
